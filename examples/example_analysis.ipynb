# Code Analysis Agent - Interactive Demo

This notebook demonstrates the multi-step reasoning capabilities of the Code Analysis & Debugging Agent.

- **Multiple Tools**: AST parsing, static analysis, git integration, code search

```python
# Setup
import sys
from pathlib import Path

# Add parent directory to path
sys.path.insert(0, str(Path.cwd().parent))

from src.agent.orchestrator import CodeAnalysisAgent
from rich.console import Console
from rich.panel import Panel
from rich.markdown import Markdown

console = Console()
```

## 1. Initialize the Agent

Point the agent to a code repository to analyze.

```python
# Initialize agent with test repository
repo_path = Path.cwd().parent / "data" / "test_repos"

agent = CodeAnalysisAgent(
    repo_path=str(repo_path),
    verbose=True  # Show reasoning steps
)

print(f"✓ Agent initialized for: {repo_path}")
print(f"✓ Available tools: {', '.join(agent.list_tools())}")
```

## 2. Basic Bug Detection

Let's analyze sample buggy code for common issues.

```python
# Analyze for bugs
result = agent.analyze(
    target="buggy_code.py",
    analysis_type="bug_detection"
)

console.print(Panel(
    result["output"],
    title="Bug Detection Results",
    border_style="green"
))
```

## 3. Understanding the Reasoning Process

The agent follows a ReAct pattern: Reason → Act → Observe → Repeat

Let's examine the reasoning steps:

```python
# Show reasoning trace
print("\n=== REASONING TRACE ===")
print(f"Total steps taken: {len(result['intermediate_steps'])}\n")

for i, (action, observation) in enumerate(result['intermediate_steps'], 1):
    console.print(f"\n[bold cyan]Step {i}: {action.tool}[/bold cyan]")
    console.print(f"[dim]Input: {action.tool_input}[/dim]")
    console.print(f"[yellow]Result:[/yellow] {str(observation)[:300]}...")
    print()
```

## 4. Security Vulnerability Scan

Now let's focus on security issues.

```python
# Security audit
security_result = agent.security_audit(target="buggy_code.py")

console.print(Panel(
    security_result["output"],
    title="Security Vulnerabilities Found",
    border_style="red"
))
```

## 5. Direct Tool Usage

You can also use individual tools directly for specific tasks.

```python
# Get AST analyzer tool
ast_tool = agent.get_tool_by_name("ast_analyzer")

# Analyze functions
function_analysis = ast_tool._run(
    '{"file_path": "buggy_code.py", "operation": "analyze_functions"}'
)

console.print(Panel(
    function_analysis,
    title="Function Analysis (AST)",
    border_style="blue"
))
```

```python
# Detect anti-patterns
pattern_analysis = ast_tool._run(
    '{"file_path": "buggy_code.py", "operation": "detect_patterns"}'
)

console.print(Panel(
    pattern_analysis,
    title="Anti-Pattern Detection",
    border_style="yellow"
))
```

## 6. Code Search

Search for specific patterns or function definitions.

```python
# Get code search tool
search_tool = agent.get_tool_by_name("code_search")

# Search for a function
search_result = search_tool._run(
    '{"query": "process_user_input", "search_type": "function_def"}'
)

print("Search Results:")
print(search_result)
```

## 7. Comprehensive Analysis

Let's run a complete analysis that checks everything.

```python
# Comprehensive analysis
comprehensive_result = agent.analyze(
    target="buggy_code.py",
    analysis_type="comprehensive"
)

console.print(Panel(
    comprehensive_result["output"],
    title="Comprehensive Analysis Report",
    border_style="green"
))
```

## 8. Summary Statistics

Let's extract some statistics from the analysis.

```python
# Count issues by severity
output = comprehensive_result["output"]

critical_count = output.lower().count("critical")
high_count = output.lower().count("high")
medium_count = output.lower().count("medium")
low_count = output.lower().count("low")

print("\n=== Issue Summary ===")
print(f"Critical: {critical_count}")
print(f"High: {high_count}")
print(f"Medium: {medium_count}")
print(f"Low: {low_count}")
print(f"\nTotal reasoning steps: {len(comprehensive_result['intermediate_steps'])}")
```
